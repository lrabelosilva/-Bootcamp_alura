## Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Sudeste e Centro-Oeste) de cada estado.

```python
usaveis["Região"] = usaveis.index.str[0]
regioes_map = {1:"Norte",2:"Nordeste",3:"Sudeste",4:"Sul",5:"Centro-Oeste"}

usaveis["Região"] = usaveis["Região"].map(lambda x: regioes_map[int(x)])

usaveis["Região"]
usaveis.head()
```
